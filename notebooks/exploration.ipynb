```python
import pandas as pd
import numpy as np
import os

try:
    df_true = pd.read_csv('../data/raw/True.csv')
    df_fake = pd.read_csv('../data/raw/Fake.csv')
    
    df_true['label'] = 0  
    df_fake['label'] = 1  
    
    df = pd.concat([df_true, df_fake], ignore_index=True)
    print("✅ Dataset berhasil digabungkan!")
    print(f"Total Berita Nyata (Real): {len(df_true)} baris")
    print(f"Total Berita Palsu (Fake): {len(df_fake)} baris")
    print(f"Total Gabungan: {len(df)} baris")
except FileNotFoundError:
    df_true = pd.read_csv('True.csv')
    df_fake = pd.read_csv('Fake.csv')
    df_true['label'] = 0
    df_fake['label'] = 1
    df = pd.concat([df_true, df_fake], ignore_index=True)
    print("✅ Dataset berhasil digabungkan dari root folder!")

df.sample(5)
```

```python
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_text(text):
    if not isinstance(text, str):
        return ""
    
    text = text.lower()
    
    text = re.sub(r'.*?\(reuters\)\s*-', '', text)
    

    text = re.sub(r'https?://\s*\S+|www\.\S+', '', text)
    
   
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    words = text.split()
    cleaned_words = [word for word in words if word not in stop_words]
    
    return " ".join(cleaned_words)

sample_index = 0
print("=== TEKS ASLI ===")
print(df['text'].iloc[sample_index][:300] + "...")
print("\n=== TEKS SETELAH DI-CLEANING ===")
print(clean_text(df['text'].iloc[sample_index])[:300] + "...")
```
